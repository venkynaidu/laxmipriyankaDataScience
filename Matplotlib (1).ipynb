```python
import numpy as np
t=np.arange(0.,5.,0.2)
plt.plot(t,t**2,'b^',label='^2')
plt.plot(t,t**2.2,'rs',label='^2.2')
plt.plot(t,t**2.5,'go',label='^2.5')
plt.legend()
plt.grid()
```
